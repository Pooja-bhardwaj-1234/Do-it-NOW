```python
pd.crosstab(df.department,df.gender).plot.bar(stacked=True)
```

```python
pd.crosstab(df.gender,df.department).plot.bar(stacked=True)
```

```python
import matplotlib.pyplot as plt
#create the crosstab
ct=pd.crosstab(df.department,df.gender)
#plot the stacked bar chart
ax=ct.plot.bar(stacked=True)
#innotate the bars with the counts
for p in ax.containers:
    ax.bar_label(p)
    
plt.show()
```
